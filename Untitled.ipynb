```python
import pandas as pd

# Assuming you've downloaded the CSV file from the provided link
file_path = "co2_emissions_data.csv"

# Try reading the file with different encodings
try:
    data = pd.read_csv(file_path, encoding='utf-8', skiprows=1)
except UnicodeDecodeError:
    data = pd.read_csv(file_path, encoding='ISO-8859-1', skiprows=1)
# or
# data = pd.read_csv(file_path, encoding='latin1')
```

```python
# Display the first few rows of the dataset
data.head()

# Get information about the dataset
data.info()

# Check for missing values
data.isnull().sum()
```

```python
# Data attributes, types, size, etc.
data.describe()
data.dtypes
data.shape
print(data.columns)

```

```python
# Filter data for Rwanda, services, and industry series
rwanda_data = data[(data['CountryName'] == 'Rwanda') & (data['Series'].isin(['Services', 'Industry']))]

# Group data by year and sum the values for each series
grouped_data = rwanda_data.groupby(['Year', 'Series'])['Value'].sum().reset_index()

# Plot trends for services and industry in Rwanda
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Value', hue='Series', data=grouped_data)
plt.title('Trends in Services and Industry in Rwanda')
plt.show()
```

```python
# Filter data for Rwanda and the two specified series
rwanda_data = data[(data['CountryName'] == 'Rwanda') & (data['Series'].isin([
    'Emissions (thousand metric tons of carbon dioxide)',
    'Emissions per capita (metric tons of carbon dioxide)'
]))]

# Group data by year and series and sum the values
grouped_data = rwanda_data.groupby(['Year', 'Series'])['Value'].sum().reset_index()

# Plot trends for the two series in Rwanda
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Value', hue='Series', data=grouped_data)
plt.title('Trends in CO2 Emissions and Emissions per Capita in Rwanda')
plt.show()
```

```python
# Filter data for Rwanda and the two specified series
rwanda_data = data[(data['CountryName'] == 'Rwanda') & (data['Series'].isin([
    'Emissions (thousand metric tons of carbon dioxide)',
    'Emissions per capita (metric tons of carbon dioxide)'
]))]

# Filter data for the last three years (2018, 2019, 2020)
rwanda_data_last_three_years = rwanda_data[rwanda_data['Year'].isin([2018, 2019, 2020])]

# Group data by year and series and sum the values
grouped_data = rwanda_data_last_three_years.groupby(['Year', 'Series'])['Value'].sum().reset_index()

# Plot trends for the two series in the last three years
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Value', hue='Series', data=grouped_data)
plt.title('Trends in CO2 Emissions and Emissions per Capita in Rwanda (2018-2020)')
plt.show()
```
