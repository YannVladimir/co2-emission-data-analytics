```python
# Question 1.a 

import pandas as pd

# Read the CSV file, skipping the first row
df = pd.read_csv("Carbon_Dioxide_Emission_Estimates.csv", encoding='latin1', skiprows=1)

# Get the column names from the DataFrame
columns = df.columns

# Print the original column names
print("Original column names:")
print(columns)
```

```python
# Rename the 'Region/Country/Area' column to 'Region/Area'
data = df.rename(columns={'Region/Country/Area': 'Region/Area'})

# Rename the 'Unnamed: 1' column to 'Country'
data = df.rename(columns={'Unnamed: 1': 'Country'})

# Print the updated column names
print("Updated column names:")

# Display the updated column names after the first renaming
print(data.columns)
```

# Question 5.a

Below we are going to understanding the data- attributes/fields, data types, data size, significant fields in our dataset

```python
print("=> Summary Statistics:")
print("-----------------------")
data.describe()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Year' and 'Value' are significant fields
significant_fields = ['Year', 'Value']

# Display basic statistics for significant fields
print("=> Basic Statistics for Significant Fields:")
print("------------------------------------------")
print(data[significant_fields].describe())

# Visualize the distribution of 'Value' over 'Year'
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Value', data=data, marker='o', color='b')
plt.title('CO2 Emission Trends Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emission Value')
plt.show()

# Visualize the distribution of 'Value' with a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['Value'], bins=20, kde=True, color='green', edgecolor='black')
plt.title('Distribution of CO2 Emission Values')
plt.xlabel('CO2 Emission Value')
plt.ylabel('Frequency')
plt.show()
```
